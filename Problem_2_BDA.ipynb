```python
import pandas as pd
import numpy as np

# Load the data

data = pd.read_excel('After_Pre_Processing.xlsx')
data = data.replace(r'^\s*$', np.nan, regex=True)



# Display the first few rows of the dataframe
print(data.head())
```

```python
inputs=data[['S1Q01nor','S1Q05','S1Q061nor']]
targets=data['S1Q06_P1nor']
    
```

```python
print(targets.unique())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
tr_S1Q01nor=LabelEncoder()
tr_S1Q05=LabelEncoder()
tr_S1Q061nor=LabelEncoder()
```

```python
inputs['S1Q01nor_n']=tr_S1Q01nor.fit_transform(inputs['S1Q01nor'])
inputs['S1Q05_n']=tr_S1Q05.fit_transform(inputs['S1Q05'])
inputs['S1Q061nor_n']=tr_S1Q061nor.fit_transform(inputs['S1Q061nor'])
```

```python
inputs.head()
```

```python
inputs_n=inputs.drop(['S1Q01nor','S1Q05','S1Q061nor'],axis='columns')
inputs_n
```

```python
tr_target = LabelEncoder()
targets_encoded = tr_target.fit_transform(targets)

# Display unique values in the transformed target variable
unique_values = set(targets_encoded)
print("Unique Values in Transformed Target Variable:", unique_values)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs_n, targets_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier(max_depth=5)
dt_classifier.fit(X_train, y_train)
```

```python
y_pred = dt_classifier.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```
